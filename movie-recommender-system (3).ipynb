# Preprocessing

```python
def dictionary_to_list(genres):
    genres = json.loads(genres)
    x = [genre["name"] for genre in genres]
    return x
```

```python
movies.genres = movies.genres.apply(dictionary_to_list)
movies.keywords = movies.keywords.apply(dictionary_to_list)
```

```python
movies.head()
```

```python
movies.cast[0]
```

```python
def cast_extractor(cast):
    cast = json.loads(cast)
    x = []
    try:
        for i in range(5):
            x.append(cast[i]["name"])
        return x
    except:
        for i in cast:
            x.append(i["name"])
        return x
            
```

```python
cast_extractor(movies.cast[0])
```

```python
movies.cast = movies.cast.apply(cast_extractor)
```

```python
def dir(c):
    for i in json.loads(c):
        if i["job"]=="Director":
            return [i["name"]]
```

```python
movies.crew = movies.crew.apply(dir)
```

```python
movies.head()
```

```python
movies = movies.dropna()
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies.genres = movies.genres.apply(" ".join)
movies.keywords = movies.keywords.apply(" ".join)
movies.cast = movies.cast.apply(" ".join)
movies.crew = movies.crew.apply(" ".join)
```

```python
movies.head()
```

```python
movies["tags"] = movies["genres"] + " " +movies["keywords"] +" " + movies["genres"] +" " + movies["cast"] +" " + movies["crew"] +" " + movies["keywords"]
```

```python
movies["tags"][0]
```

```python
movies.drop(['genres', 'keywords', 'overview','cast' ,'crew'], axis = 1, inplace = True)
```

```python
movies["original_title"]=movies["original_title"].apply(lambda x: x.lower())
movies["tags"]=movies["tags"].apply(lambda x: x.lower())
```

```python
movies.head()
```

# Stemming vectorization cosine similarities

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemer(x):
    y  =[]
    for i in x.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies.tags=movies.tags.apply(stemer)
```

```python
movies.tags[95]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer( max_features= 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(movies['tags'])
```

```python
cv.get_feature_names_out()[0]
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cs = cosine_similarity(vectors)
```

```python
cs[0:2]
```

# Recommend

```python
def recommend(searched_movie):
    ## get movie name 
    
    ## get the index of the movie
    searched_movie_index = movies[movies['original_title']==searched_movie.lower()].index[0]
    
    ## search the most similar in the cosine similaritues
    most_similar_movies = sorted(list(enumerate(cs[searched_movie_index])), reverse = True, key = lambda x: x[1])[1:6]

    ## get the movies by index
    l = []
    for i in most_similar_movies:
        l.append(movies.iloc[i[0]]['original_title'].title())
    

    return l
```

```python
recommend("batman")
```
